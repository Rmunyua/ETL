```python
import pandas as pd
```

```python
url = 'https://www.bls.gov/web/metro/laummtrk.htm#laummtrk.f.p'
```

```python
tables = pd.read_html(url)
tables
```

```python
df = tables[0]
df.columns = ['Metropolitan area', 'Unemployment Rate', 'Rank']
df.dropna(inplace=True)
df2 = df.iloc[:389,]
df2["Year"] = 2020
df2["Month"] = 3
df2
```

Pandas also had a `to_html` method that we can use to generate HTML tables from DataFrames.

```python
new = df2["Metropolitan area"].str.split(", ", n = 1, expand = True)
df2["City/Area"] = new[0]
new2 = new[1].str.split(" ", n = 1, expand = True)
df2["State"] = new2[0]
df2
```

```python
df2.to_csv("unemployment.csv", index=False, header=True)
```
